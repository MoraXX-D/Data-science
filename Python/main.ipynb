```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
print(df.describe())
print(df.info())
print(df.isnull().sum())
```

# Droping unnamed coloumn

```python
df = df.drop("Unnamed: 0",axis =1)
print(df.head())
```

# Gender distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

# analysis : Number of women > Number of men

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Parent Education's impact")
plt.show()
```

# analysis : Parent's education is directly proportional to the performance of education in exams, that is higher the parent's education higher the student number

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Parent marital status impact")
plt.show()
```

# Analysis : There is no/negligible impact on the performance of student based on their parent's marital status

# Distribution of Ethenic group

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()
# print(groupA,groupB,groupC,groupD,groupE)

mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
label = ["groupA", "groupB", "groupC", "groupD", "groupE"]
print(mlist)
plt.pie(mlist, labels=label,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show()
```
